```python
# Splitting data in to X, Y for training and for testing the ML model.

y=df['bwt']
x = df.iloc[:,1:]
x.head()
```

```python
# train test split. (we use scikit learn lib in python.)
from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train, y_test = tts(x,y,test_size=0.20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Lasso,Ridge
```

```python
lr = LinearRegression()
lasso = Lasso()
ridge = Ridge ()
```

```python
# Training the model
lr.fit(x_train,y_train)
lasso.fit(x_train,y_train)
ridge.fit(x_train,y_train)
```

```python
# Predicting.
lr_pred = lr.predict(x_test)
las_pred = lasso.predict(x_test)
rid_pred = ridge.predict(x_test)
```

```python
# Model evaluation.

from sklearn.metrics import r2_score,mean_squared_error
```

```python
lr_rscore= r2_score(y_test,lr_pred)
las_rscore= r2_score(y_test,las_pred)
rid_rscore= r2_score(y_test,rid_pred)
```

```python
lr_mse= mean_squared_error(y_test,lr_pred)
las_mse= mean_squared_error(y_test,las_pred)
rid_mse= mean_squared_error(y_test,rid_pred)
```

```python
print("\n performance of the Linear Regration Model :" )
print("R2 Score :",lr_rscore)
print("MSE Score :",lr_mse)


print("\n performance of the Lasso Regration Model :" )
print("R2 Score :",las_rscore)
print("MSE Score :",las_mse)


print("\n performance of the Ridge Regration Model :" )
print("R2 Score :",rid_rscore)
print("MSE Score :",rid_mse)
```
